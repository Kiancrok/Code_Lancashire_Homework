```python
f1_df = pd.read_csv(r'f1\F1Drivers_Dataset.csv', sep=',')
```

```python
f1_df.head()
```

```python
schumacher = f1_df[f1_df['Driver'] == 'Michael Schumacher']
senna = f1_df[f1_df['Driver'] == 'Ayrton Senna']
```

```python
schumacherStats = schumacher[['Race_Wins','Pole_Positions','Podiums','Fastest_Laps','Championships','Win_Rate','Pole_Rate','Podium_Rate','FastLap_Rate']]
sennaStats = senna[['Race_Wins','Pole_Positions','Podiums','Fastest_Laps','Championships','Win_Rate','Pole_Rate','Podium_Rate','FastLap_Rate']]
print("Michael Schumacher's Stats:")
print(schumacherStats)
print("\nAyrton Senna's Stats:")
print(sennaStats)
```

```python
schumacherStats = np.array([[ 91.0,68.0,155.0,77.0,7.0,0.295455 ,  0.220779  ,   0.503247    ,      0.25  
]]) 
sennaStats = np.array([[ 41.0   ,         65.0 ,    80.0     ,     19.0      ,      3.0,  0.253086  , 0.401235 ,    0.493827  ,    0.117284  ]]) 

# Statistics names
stats_names = ['Race_Wins', 'Pole_Positions', 'Podiums', 'Fastest_Laps', 'Championships', 'Win_Rate', 'Pole_Rate', 'Podium_Rate', 'FastLap_Rate']

# Plot
plt.figure(figsize=(12, 6))

# Plot each driver's statistics
plt.plot(np.arange(len(stats_names)), schumacherStats[0], label='Michael Schumacher', linestyle='-', marker='o', color='red')
plt.plot(np.arange(len(stats_names)), sennaStats[0], label='Ayrton Senna', linestyle='-', marker='o', color='green' )

# Bottom row for all the stats
plt.xticks(np.arange(len(stats_names)), stats_names, rotation=45, ha='right')

# Legend
plt.legend()

# Labels and title
plt.xlabel('Statistics')
plt.ylabel('Values')
plt.title('Comparison of Racing Statistics: Michael Schumacher vs Ayrton Senna')

# Show plot
plt.tight_layout()
plt.show()


```

```python
data = pd.read_csv('f1/F1Drivers_Dataset.csv')

# Define statistics names
stats_names = ['Race_Wins', 'Pole_Positions', 'Podiums', 'Fastest_Laps', 'Championships']

# Extract statistics for Michael Schumacher and Ayrton Senna
schumacherStats = data[data['Driver'] == 'Michael Schumacher'][stats_names].values
sennaStats = data[data['Driver'] == 'Ayrton Senna'][stats_names].values

# Plot
plt.figure(figsize=(10, 6))

# Plot each driver's statistics
plt.plot(np.arange(len(stats_names)), schumacherStats[0], label='Michael Schumacher', linestyle='-', marker='o', color='red')
plt.plot(np.arange(len(stats_names)), sennaStats[0], label='Ayrton Senna', linestyle='-', marker='o', color='green')

# Bottom row for all the stats
plt.xticks(np.arange(len(stats_names)), stats_names, rotation=0, ha='right')

# Legend
plt.legend()

# Labels and title
plt.xlabel('Statistics')
plt.ylabel('Values')
plt.title('Why Michael Shumacher Is Better Than Ayrton Senna')

# Show plot
plt.tight_layout()
plt.show()
```
